```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
import pandas as pd
import graph_tool as gt
from graph_tool import centrality, inference
from tqdm import tqdm
from operator import itemgetter
from functools import reduce
import itertools
import networkx as nx
#from py3plex.core import multinet
#from py3plex.core import random_generators

import warnings
warnings.filterwarnings("ignore")


#set.seed(1)

# input data settings
NEIGH_ORDER = 1 # or 0, order of nerighbors, 0 only connected proteins, 1 also first neighbors
CUT_THR = 0.7   # don't change this one

# multilayer settings
layerCouplingStrength = 1
networkOfLayersType = "categorical" ## = all to all

#virus metadata
virus_metadata = pd.read_csv("Files/viruses_metadata.csv", header=0, sep=";")

virus_metadata_onco = virus_metadata[virus_metadata["isOncogenic"] == True].reset_index()
virus_metadata_nonco = virus_metadata[virus_metadata["isOncogenic"] == False].reset_index()

#dictionary containing a unquie mapping between name of the protein and a corresponding index
node_map_df = pd.read_csv("Files/node_map.csv")
node_map_dict = {k:(v-1) for k,v in zip(node_map_df["Prot"], node_map_df["Index"])}

#function to create list of n_iter combination of nonco virus indexes with a fixed random seed for repitibility
def SamplingForNoco(n_iter, start=0, group_dim=8):
    np.random.seed(1234)
    nonco_cond = np.where(np.all([np.array(virus_metadata["virus"]!="Human_SARS_coronavirus_2"),
                                  np.array(virus_metadata["neigh_order"]==NEIGH_ORDER), 
                                  np.array(virus_metadata["isOncogenic"]==False)],
                                  axis=0))
    
    nonco_sampling = np.array([np.random.choice(nonco_cond[0], group_dim, replace=False) for i in range(n_iter+start)])
    
    return nonco_sampling[start:(n_iter+start)]

def writeComponent(fname, ensemble):
    with open(fname, "w") as cio:
        for i in range(len(ensemble)):
            cio.write(" ".join(map(str,ensemble[i]))+"\n")

def readComponent(fname):
    with open(fname, "r") as fread:
        read_list = fread.readlines()
    var = []
    for i in range(len(read_list)):
        if read_list[i][:-1]=="":
            var.append(np.array([]))
        else:
            var.append(np.array(list(map(int,read_list[i][:-1].split(" ")))))
    return var
```

```python
class VirusMultiplex():
    def __init__(self, indexes):
        self.target_folder = "Virus_data_Enriched_"+str(CUT_THR)+"_Neigh_"+str(NEIGH_ORDER)+"/"
        self.indexes = indexes
        self.mux_ppi = pd.DataFrame()
        self.virus_list = []
        for i in self.indexes:
            if virus_metadata.loc[i,"neigh_order"]==NEIGH_ORDER:
                virus_path = self.target_folder + virus_metadata.loc[i,"virus"]
                self.virus_list.append(virus_metadata.loc[i,"virus_short"])

                virus_edges = pd.read_csv(virus_path+"/edges.csv", header=0, sep=",")
                virus_nodes = pd.read_csv(virus_path+"/nodes.csv", header=0, sep=",")

                if virus_nodes.shape[0]<=1: continue
                
                human_prot = virus_nodes[np.any([(virus_nodes["type"]==1), (virus_nodes["type"]==2)], axis=0)]["node"]
                human_ppi = virus_edges[np.all([(virus_edges["V1"].isin(human_prot)), (virus_edges["V2"].isin(human_prot))],axis=0)]

                if human_ppi.shape[0]>0:
                    human_ppi["layer"] = virus_metadata.loc[i,"virus_short"]
                    self.mux_ppi = pd.concat([self.mux_ppi, human_ppi])

        self.mux_ppi = self.mux_ppi.reset_index()


        self.Layers = self.mux_ppi["layer"].unique().shape[0]
        self.layer_map = {self.mux_ppi["layer"].unique()[i]: i for i in range(self.Layers) }

        # mapping node names to ids
        self.unique_nodes = pd.concat([self.mux_ppi["V1"], self.mux_ppi["V2"]]).unique()
        self.node_map = {self.unique_nodes[i]: i for i in range(self.unique_nodes.shape[0]) }

        self.Nodes = self.unique_nodes.shape[0]
        self.Edges = self.mux_ppi.shape[0]

        self.mux_ppi["nodeA"] = itemgetter(*self.mux_ppi["V1"].to_numpy())(self.node_map)
        self.mux_ppi["nodeB"] = itemgetter(*self.mux_ppi["V2"].to_numpy())(self.node_map)
        self.mux_ppi["l"] = itemgetter(*self.mux_ppi["layer"].to_numpy())(self.layer_map)
        
        #generate an list of networks and aggregate/aggregate-binary network
        self.Tensor_onco = []
        self.g_list = []
        self.M_agg = np.zeros([self.Nodes, self.Nodes])
        
        for l in range(self.Layers):
            tmp = self.mux_ppi[self.mux_ppi["l"]==l].reset_index()[["nodeA", "nodeB"]]
            tmp = tmp.rename({"nodeA":"source", "nodeB":"target"}, axis=1)
            self.g_list.append(gt.Graph(directed=False))
            self.g_list[l].add_edge_list(tmp.values)
            self.g_list[l].add_vertex(self.Nodes-self.g_list[l].num_vertices())
            self.M_agg = self.M_agg + gt.spectral.adjacency(self.g_list[l])/2


        self.M_agg_binary = np.copy(self.M_agg)
        self.M_agg_binary[self.M_agg_binary>=1]=1

        self.g_agg = gt.Graph(directed=False)
        self.g_agg.add_edge_list(np.transpose(self.M_agg_binary.nonzero()))
        
        self.net_description = "Multiplex with " + str(self.Layers) + " layers, " + str(self.Nodes) + " nodes and " + str(self.Edges) + " edges"
        
        #generate the mulilayer network structure
        tmp_multi = self.mux_ppi.reset_index()[["nodeA", "nodeB"]]
        tmp_multi = tmp_multi.rename({"nodeA":"source", "nodeB":"target"}, axis=1)
        self.g_multi = gt.Graph(directed=False)
        self.g_multi.add_edge_list(tmp.values)
        ep_weight_lay = []
        
        for l in range(self.Layers):
            ep_weight_lay.append([l]*self.mux_ppi["l"].value_counts()[l])

        ep_weight_lay = np.concatenate(ep_weight_lay)
        self.g_multi.edge_properties["weight"]=self.g_multi.new_edge_property("int", ep_weight_lay)
        
    def GetLCC(self):
        lcc = np.array(itemgetter(*np.array(list(self.node_map.keys()))[gt.topology.extract_largest_component(self.g_agg).get_vertices()])(node_map_dict))
        return lcc

    def GetLIC(self):
        lcc_per_layer = [gt.topology.extract_largest_component(self.g_list[i]).get_vertices() for i in range(self.Layers)]
        lcc_inters = reduce(np.intersect1d, lcc_per_layer)
        lic = np.array(itemgetter(*np.array(list(self.node_map.keys()))[lcc_inters])(node_map_dict))
        return lic

    def GetMod(self, n_iter):
        self.modules_list = []
        self.modularity_list = []

        for It_Com in tqdm(range(n_iter)):
            self.state_multi = gt.inference.minimize_blockmodel_dl(self.g_multi,
                                                              state_args=dict(base_type=gt.inference.LayeredBlockState,
                                                              state_args=dict(ec=self.g_multi.ep.weight, layers=True)))

            self.modules_list.append(self.state_multi.get_nonempty_B())
            self.modularity_list.append(inference.modularity(self.g_multi, self.state_multi.get_blocks()))
    
        return self.modules_list, self.modularity_list 

    def percolation(self):
        self.perc_agg_1 = gt.topology.vertex_percolation(self.g_agg, np.argsort(gt.centrality.pagerank(self.g_agg).get_array()))[0]
        self.perc_agg_2 = gt.topology.vertex_percolation(self.g_agg, np.argsort(gt.centrality.pagerank(self.g_agg).get_array()), second=True)[0]
        self.max_perc = np.argmax(self.perc_agg_2)/len(self.perc_agg_1)

        return self.max_perc
                
```

```python
def createNet(index):
    target_folder = "Virus_data_Enriched_"+str(CUT_THR)+"_Neigh_"+str(NEIGH_ORDER)+"/"
    mux_ppi = pd.DataFrame()
    virus_list = []
    if virus_metadata.loc[index,"neigh_order"]==NEIGH_ORDER:
        virus_path = target_folder + virus_metadata.loc[index,"virus"]

        virus_edges = pd.read_csv(virus_path+"/edges.csv", header=0, sep=",")
        virus_nodes = pd.read_csv(virus_path+"/nodes.csv", header=0, sep=",")
        
        human_prot = virus_nodes[np.any([(virus_nodes["type"]==1), (virus_nodes["type"]==2)], axis=0)]["node"]
        human_ppi = virus_edges[np.all([(virus_edges["V1"].isin(human_prot)), (virus_edges["V2"].isin(human_prot))],axis=0)]

        if human_ppi.shape[0]>0:
            mux_ppi = human_ppi

    mux_ppi = mux_ppi.reset_index()


    # mapping node names to ids
    unique_nodes = pd.concat([mux_ppi["V1"], mux_ppi["V2"]]).unique()
    node_map = {unique_nodes[i]: i for i in range(unique_nodes.shape[0]) }

    Nodes = unique_nodes.shape[0]
    Edges = mux_ppi.shape[0]

    mux_ppi["nodeA"] = itemgetter(*mux_ppi["V1"].to_numpy())(node_map)
    mux_ppi["nodeB"] = itemgetter(*mux_ppi["V2"].to_numpy())(node_map)
    

    tmp = mux_ppi[["nodeA", "nodeB"]]
    tmp = tmp.rename({"nodeA":"source", "nodeB":"target"}, axis=1)
    g = gt.Graph(directed=False)
    g.add_edge_list(tmp.values)

    return g
```
