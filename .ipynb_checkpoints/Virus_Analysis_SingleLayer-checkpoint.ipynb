# Libraries

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
import pandas as pd
import graph_tool as gt
from graph_tool import centrality, inference
from tqdm import tqdm
from operator import itemgetter
from functools import reduce
import itertools
#from pymnet import *

import warnings
warnings.filterwarnings("ignore")


#set.seed(1)

# input data settings
NEIGH_ORDER = 1 # or 0, order of nerighbors, 0 only connected proteins, 1 also first neighbors
CUT_THR = 0.7   # don't change this one

# multilayer settings
layerCouplingStrength = 1
networkOfLayersType = "categorical" ## = all to all

#virus metadata
virus_metadata = pd.read_csv("Files/viruses_metadata.csv", header=0, sep=";")

virus_metadata_onco = virus_metadata[virus_metadata["isOncogenic"] == True].reset_index()
virus_metadata_nonco = virus_metadata[virus_metadata["isOncogenic"] == False].reset_index()

#dictionary containing a unquie mapping between name of the protein and a corresponding index
node_map_df = pd.read_csv("Files/node_map.csv")
node_map_dict = {k:(v-1) for k,v in zip(node_map_df["Prot"], node_map_df["Index"])}

#function to create list of n_iter combination of nonco virus indexes with a fixed random seed for repitibility
def SamplingForNoco(n_iter, start=0, group_dim=8):
    np.random.seed(1234)
    nonco_cond = np.where(np.all([np.array(virus_metadata["virus"]!="Human_SARS_coronavirus_2"),
                                  np.array(virus_metadata["neigh_order"]==NEIGH_ORDER), 
                                  np.array(virus_metadata["isOncogenic"]==False)],
                                  axis=0))
    
    nonco_sampling = np.array([np.random.choice(nonco_cond[0], group_dim, replace=False) for i in range(n_iter+start)])
    
    return nonco_sampling[start:(n_iter+start)]

def writeComponent(fname, ensemble):
    with open(fname, "w") as cio:
        for i in range(len(ensemble)):
            cio.write(" ".join(map(str,ensemble[i]))+"\n")

def readComponent(fname):
    with open(fname, "r") as fread:
        read_list = fread.readlines()
    var = []
    for i in range(len(read_list)):
        if read_list[i][:-1]=="":
            var.append(np.array([]))
        else:
            var.append(np.array(list(map(int,read_list[i][:-1].split(" ")))))
    return var
```

```python
onco_virus_indexes = np.where(np.array(np.all([virus_metadata["neigh_order"]==NEIGH_ORDER, 
                                               virus_metadata["isOncogenic"] == True], 
                                              axis=0)))[0]

nonco_virus_indexes= np.where(np.all([np.array(virus_metadata["virus"]!="Human_SARS_coronavirus_2"),
                                      np.array(virus_metadata["neigh_order"]==NEIGH_ORDER), 
                                      np.array(virus_metadata["isOncogenic"]==False)],
                                      axis=0))[0]
nonco_virus_indexes=list(nonco_virus_indexes)
nonco_virus_indexes.remove(98)
nonco_virus_indexes=np.array(nonco_virus_indexes)
```

# Single Layer analysis

```python
class VirusNet():
    def __init__(self, index):
        target_folder = "Virus_data_Enriched_"+str(CUT_THR)+"_Neigh_"+str(NEIGH_ORDER)+"/"
        self.mux_ppi = pd.DataFrame()
        
        if virus_metadata.loc[index,"neigh_order"]==NEIGH_ORDER:
            virus_path = target_folder + virus_metadata.loc[index,"virus"]

            virus_edges = pd.read_csv(virus_path+"/edges.csv", header=0, sep=",")
            virus_nodes = pd.read_csv(virus_path+"/nodes.csv", header=0, sep=",")
            
            human_prot = virus_nodes[np.any([(virus_nodes["type"]==1), (virus_nodes["type"]==2)], axis=0)]["node"]
            human_ppi = virus_edges[np.all([(virus_edges["V1"].isin(human_prot)), (virus_edges["V2"].isin(human_prot))],axis=0)]

            if human_ppi.shape[0]>0:
                self.mux_ppi = human_ppi

        self.mux_ppi = self.mux_ppi.reset_index()


        # mapping node names to ids
        unique_nodes = pd.concat([self.mux_ppi["V1"], self.mux_ppi["V2"]]).unique()
        self.node_map = {unique_nodes[i]: i for i in range(unique_nodes.shape[0]) }

        Nodes = unique_nodes.shape[0]
        Edges = self.mux_ppi.shape[0]

        self.mux_ppi["nodeA"] = itemgetter(*self.mux_ppi["V1"].to_numpy())(self.node_map)
        self.mux_ppi["nodeB"] = itemgetter(*self.mux_ppi["V2"].to_numpy())(self.node_map)
        

        tmp = self.mux_ppi[["nodeA", "nodeB"]]
        tmp = tmp.rename({"nodeA":"source", "nodeB":"target"}, axis=1)
        self.g = gt.Graph(directed=False)
        self.g.add_edge_list(tmp.values)
```

## Degree distribution

```python
from scipy.optimize import curve_fit

def f(x, A, k, expo):
    return A+k*np.power(x, expo)
```

### Onco

```python
degree_list = []
for ind in onco_virus_indexes:
    net = VirusNet(ind)
    degree_list.append(net.g.get_total_degrees(np.arange(net.g.num_vertices())))
```

```python
writeComponent("Files/SingleLayer/degrees/oncogenic.txt", degree_list)
```

```python
degree_list = readComponent("Files/SingleLayer/degrees/oncogenic.txt")
```

```python
expo_list_onco=[]
_, ax = plt.subplots(2,4, figsize=(12,8))
ax = ax.flatten()
for i in range(len(degree_list)):
    deg,cnt = np.unique(degree_list[i], return_counts=True)
    res=curve_fit(f=f, xdata=deg[5:400], ydata=cnt[5:400], p0=(200,100,-0.12), maxfev=5000)
    expo_list_onco.append(res[0][-1])
    ax[i].plot(deg[5:400], cnt[5:400])
    ax[i].plot(np.arange(5,400), f(np.arange(5,400), *res[0]), label="expo %.5f"%res[0][-1])
    ax[i].legend()
```

```python
np.savetxt(X=expo_list_onco, fname="Files/SingleLayer/degrees/onco_expo.txt")
```

## Betweenness

### Onco

```python
# onco nets
node_map_list= []
betw_list=[]
net_list=[]
for ind in tqdm(onco_virus_indexes):
    net = VirusNet(ind)
    node_map_list.append(net.node_map)
    betw_list.append(gt.centrality.betweenness(net.g)[0].get_array())
```

```python
#write
for j in range(len(betw_list)):
    nm = list(node_map_list[j].keys())
    name = "Files/SingleLayer/betweenness/"+virus_metadata.iloc[onco_virus_indexes[j],:].virus_short+".txt"
    print(name)
    pd.DataFrame({"Prot": nm, "Betweeness": betw_list[j]}).to_csv(name)
```

```python
#read
betw_list=[]
for ind in onco_virus_indexes:
    net = VirusNet(ind)
    nm = list(net.node_map.keys())
    name = "Files/SingleLayer/betweenness/"+virus_metadata.iloc[ind,:].virus_short+".txt"
    print(name)
    betw_list.append(np.array(pd.read_csv(name)["Betweeness"]))
```

```python
max_perc_list_onco=[]
for i in tqdm(range(len(onco_virus_indexes))):
    net = VirusNet(onco_virus_indexes[i])
    perc_1 = gt.topology.vertex_percolation(net.g, np.argsort(betw_list[i]))[0]
    perc_2 = gt.topology.vertex_percolation(net.g, np.argsort(betw_list[i]), second=True)[0]
    max_perc_list_onco.append(np.argmax(perc_2)/len(perc_1))
```

```python
np.savetxt(X=max_perc_list_onco, fname="Files/SingleLayer/percolation/onco_crit_point_betw.txt")
```
