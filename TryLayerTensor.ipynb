# Layer trials

```python
# input data settings
NEIGH_ORDER = 1 # or 0, order of nerighbors, 0 only connected proteins, 1 also first neighbors
CUT_THR = 0.7   # don't change this one

target_folder = "Virus_data_Enriched_"+str(CUT_THR)+"_Neigh_"+str(NEIGH_ORDER)+"/"

# multilayer settings
layerCouplingStrength = 1
networkOfLayersType = "categorical" ## = all to all

#virus metadata
virus_metadata = pd.read_csv("Files/viruses_metadata.csv", header=0, sep=";")

virus_metadata_onco = virus_metadata[virus_metadata["isOncogenic"] == True].reset_index()
virus_metadata_nonco = virus_metadata[virus_metadata["isOncogenic"] == False].reset_index()

#dictionary containing a unquie mapping between name of the protein and a corresponding index
node_map_df = pd.read_csv("Files/node_map.csv")
node_map_dict = {k:(v-1) for k,v in zip(node_map_df["Prot"], node_map_df["Index"])}

onco_virus_indexes = np.where(np.array(np.all([virus_metadata["neigh_order"]==NEIGH_ORDER, virus_metadata["isOncogenic"] == True], axis=0)))[0]
```

```python
net = mxp.VirusMultiplex(onco_virus_indexes, target_folder=target_folder, virus_metadata=virus_metadata)
```

```python
node_tensor = mxp.build.get_node_tensor_from_network_list(net.g_list)
g_agg = mxp.build.get_aggregate_network(obj=node_tensor)
M_agg = mxp.build.get_node_tensor_from_network_list([g_agg])[0]
```

```python
def build_cust_LT(vec):
    splitted = np.split(vec,[7,7+6,7+6+5,7+6+5+4,7+6+5+4+3,7+6+5+4+3+2,7+6+5+4+3+2+1])
    h_mat = []
    for i in range(8):
        l1=np.zeros(8)
        l1[i+1:]=splitted[i]
        h_mat.append(l1)
    
    h_mat = np.array(h_mat)
    
    layer_tensor = h_mat+h_mat.T
    return layer_tensor

def get_std(vec, node_tensor, layer_tensor, net, retlist=False):
    layer_tensor = build_cust_LT(vec)
    
    supra = mxp.build.build_supra_adjacency_matrix_from_edge_colored_matrices(nodes_tensor=node_tensor,
                                                                              layer_tensor=layer_tensor,
                                                                              layers=net.Layers,
                                                                              nodes=net.Nodes)

    kos=mxp.versatility.get_multi_RW_centrality(supra, net.Layers, net.Nodes, Type="pagerank", multilayer=False)

    occ_list = []
    for i in range(net.Layers-1):
        occ_list.append(kos[net.Nodes*i:net.Nodes*(i+1)].sum())
    
    if retlist:
        return occ_list
    return np.std(occ_list)
```

```python
def objective(trial):

    # set the range of the transformations to apply to the image
    hyperparameters = [
        trial.suggest_int("l01", 1, 100),
        trial.suggest_int("l02", 1, 100),
        trial.suggest_int("l03", 1, 100),
        trial.suggest_int("l04", 1, 100),
        trial.suggest_int("l05", 1, 100),
        trial.suggest_int("l06", 1, 100),
        trial.suggest_int("l07", 1, 100),
        trial.suggest_int("l12", 1, 100),
        trial.suggest_int("l13", 1, 100),
        trial.suggest_int("l14", 1, 100),
        trial.suggest_int("l15", 1, 100),
        trial.suggest_int("l16", 1, 100),
        trial.suggest_int("l17", 1, 100),
        trial.suggest_int("l23", 1, 100),
        trial.suggest_int("l24", 1, 100),
        trial.suggest_int("l25", 1, 100),
        trial.suggest_int("l26", 1, 100),
        trial.suggest_int("l27", 1, 100),
        trial.suggest_int("l34", 1, 100),
        trial.suggest_int("l35", 1, 100),
        trial.suggest_int("l36", 1, 100),
        trial.suggest_int("l37", 1, 100),
        trial.suggest_int("l45", 1, 100),
        trial.suggest_int("l46", 1, 100),
        trial.suggest_int("l47", 1, 100),
        trial.suggest_int("l56", 1, 100),
        trial.suggest_int("l57", 1, 100),
        trial.suggest_int("l67", 1, 100)
    ]
    
    return get_std(hyperparameters, node_tensor, layer_tensor, net)

pruner: optuna.pruners.BasePruner = optuna.pruners.NopPruner()
sampler = TPESampler(seed=12345)
study = optuna.create_study(study_name="Pagarank", direction="minimize", pruner=pruner, sampler=sampler)
optuna.logging.set_verbosity(optuna.logging.WARNING)
study.optimize(objective, n_trials=50)


print("Number of finished trials: {}".format(len(study.trials)))

print("Best trial:")
trial = study.best_trial

print("   Max IoU: {}".format(trial.value))

print("   Params: ")
for key, value in trial.params.items():
    print("    {}: {}".format(key, value))
```

```python
final_LT = build_cust_LT(np.array(list(trial.params.values())))
```
